```python
import sqlite3
import pandas as pd 
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
# Create your connection.
df1 = sqlite3.connect('database.sqlite')
df = pd.read_sql_query("SELECT * FROM Player_Attributes", df1)
```

```python
features = [
       'potential', 'crossing', 'finishing', 'heading_accuracy',
       'short_passing', 'volleys', 'dribbling', 'curve', 'free_kick_accuracy',
       'long_passing', 'ball_control', 'acceleration', 'sprint_speed',
       'agility', 'reactions', 'balance', 'shot_power', 'jumping', 'stamina',
       'strength', 'long_shots', 'aggression', 'interceptions', 'positioning',
       'vision', 'penalties', 'marking', 'standing_tackle', 'sliding_tackle',
       'gk_diving', 'gk_handling', 'gk_kicking', 'gk_positioning',
       'gk_reflexes']
```

```python
target = ['overall_rating']
```

```python

df = df.dropna()
```

```python
X = df[features]
```

```python
y=df[target]
```

```python
y
```

```python
y.max()
```

```python
sns.scatterplot(x='gk_reflexes',y='overall_rating',data=df,hue='preferred_foot')
```

```python
sns.relplot(x='gk_reflexes',y='overall_rating',data=df,hue='preferred_foot')
```

```python
sns.lineplot(x='gk_reflexes',y='overall_rating',data=df,hue='preferred_foot')
```

```python


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=324)
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_prediction = regressor.predict(X_test)
y_prediction
```

```python
y_test.describe()
```

```python

RMSE = sqrt(mean_squared_error(y_true = y_test, y_pred = y_prediction))
```

```python

print(RMSE)
```

```python
regressor = DecisionTreeRegressor(max_depth=20)
regressor.fit(X_train, y_train)
```

```python
y_prediction = regressor.predict(X_test)
y_prediction
```

```python

y_test.describe()
```

```python
RMSE = sqrt(mean_squared_error(y_true = y_test, y_pred = y_prediction))
```
